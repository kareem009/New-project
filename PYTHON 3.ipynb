```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
from math import pi
```

```python
df = pd.read_csv("CompleteDataset.csv")
#get columns name
a=df.columns
# append columns name in a list 
col_name=list(a)
# to get the best striker filtered
Best_Strikers=(df["Age"] <= 27) & ((df["ST"] > 70) | (df["RS"] > 70) | (df["LS"] > 70)| (df["Overall"] > 90)) 
# assign the filtered view in a new varible
filtered_view=df[Best_Strikers] 
#put the values of the filtered view in a matrix
new_df=filtered_view.iloc[0:5,:].values 
#assign the new matrix in a data frame
x=pd.DataFrame(new_df,columns=col_name)
#take the top 5 values 
top_5=x.loc[:5,["Name", "Age","Club","Overall","Value","Acceleration","Aggression","Agility","Balance","Ball control","Composure","Crossing","Curve","Dribbling","Finishing","Free kick accuracy"]]
#first player
new_df2=x.iloc[0:1,:].values 
first_player=pd.DataFrame(new_df2,columns=col_name)
first_player.iloc[0:5,:]

#RADAR CHART
 
# Set data
df1 = pd.DataFrame({
'group': ['A'],
'Acceleration':[50] ,
'Agility': [90],
'Ball control': [90],
'Curve': [90],
'Dribbling': [90]
})
 
# number of variable
categories=list(df1)[1:]
N = len(categories)
 
# We are going to plot the first line of the data frame.
# But we need to repeat the first value to close the circular graph:
values=df1.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
values
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories, color='black', size=10)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks(range(0,100,10), color="white", size=7)
plt.ylim(0,100)
 
# Plot data
ax.plot(angles, values, linewidth=1, linestyle='solid')
 
# Fill area
ax.fill(angles, values, 'B', alpha=0.1)
```
